#### Section 4: Nested Loops

##### Problem 4.1: Pattern Printing - Right Triangle (Easy)

Print the following pattern for n rows:

```
*
**
***
****
*****
```
(Example shown for n=5)

**Hint:** Use nested loops: outer loop for rows (i from 1 to n), inner loop for columns (j from 1 to i). Print * in inner loop.

````python
n = int(input())
for i in range(1, n+1):
    for j in range(i):
        print('*', end=' ')
    print()
        
````

##### Problem 4.2: Pattern Printing - Inverted Right Triangle (Easy)

Print the following pattern for n rows:

```
*****
****
***
**
*
```
(Example shown for n=5)

**Hint:** Outer loop for rows. Inner loop prints stars from 1 to (n-i+1), where i is the row number.

````python
n = int(input())
for i in range(1, n+1):
    for j in range(n-i+1):
        print('*', end=' ')
    print()
````

### Problem 4.3: Pattern Printing - Pyramid (Medium)

Print the following pyramid pattern for n rows:

```
    *
   ***
  *****
 *******
*********
```
(Example shown for n=5)

**Hint:** For row i: print (n-i) spaces, then print (2*i-1) stars. Use nested loops or string multiplication.

````python
n = int(input())
for i in range(1, n+1): #print rows
    for j in range(n-i): #print spaces
        print(' ', end='')
    for k in range(2*i-1): #print stars
        print('*', end='')
    print()
````

### Problem 4.4: Pattern Printing - Number Pattern (Medium)

Print the following number pattern for n rows:

```
1
12
123
1234
12345
```
(Example shown for n=5)

**Hint:** Similar to Problem 4.1, but print j (column number) instead of *.

````python
n = int(input())
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end='')
    print()
````

##### Problem 4.5: Multiplication Table Grid (Medium)

Print a multiplication table grid from 1 to n.

Example for n=5:
```
1   2   3   4   5
2   4   6   8  10
3   6   9  12  15
4   8  12  16  20
5  10  15  20  25
```

**Hint:** Nested loops: for i in range(1, n+1) and for j in range(1, n+1). Print i*j with proper spacing.

````python
n= int(input())
for i in range(1, n+1):
    for j in range(1, n+1):
        print(f'{i*j:4}', end='')
    print()
````

##### Problem 4.6: Diamond Pattern (Hard)

Print a diamond pattern for n rows (n should be odd):

```
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```
(Example shown for n=9)

**Hint:** Print pyramid (upper half) for rows 1 to (n+1)//2, then print inverted pyramid (lower half) for remaining rows.

````python
n = int(input())
# upper including middle line
for i in range(1, (n+1)//2+1):
    for s in range((n//2)-(i-1)):
        print(' ', end='')
    for j in range(2*i-1):
        print('*', end='')
    print()
# lower
for i in range((n-1)//2, 0, -1):
    for s in range((n//2)-(i-1)):
        print(' ', end='')
    for j in range(2*i-1):
        print('*', end='')
    print()
    
````
